```python
# prompt: check if the Profile URL has duplicates

duplicate_profiles = df[df['Profile URL'].duplicated(keep=False)]
if not duplicate_profiles.empty:
  print("Duplicate Profile URLs found:")
  # Convert the 'Profile URL' column to a list using the tolist() method of the Series
  print(duplicate_profiles['Profile URL'])
else:
  print("No duplicate Profile URLs found.")
```

```python
therapist_data=pd.read_csv("/content/therapist_data.csv")

duplicate_profiles = therapist_data[therapist_data['Profile URL'].duplicated(keep=False)]
if not duplicate_profiles.empty:
  print("Duplicate Profile URLs found:")
  # Convert the 'Profile URL' column to a list using the tolist() method of the Series
  print(duplicate_profiles['Profile URL'])
else:
  print("No duplicate Profile URLs found.")
```
